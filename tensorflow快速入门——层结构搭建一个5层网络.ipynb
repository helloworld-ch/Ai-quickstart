```python
import tensorflow as tf
```

```python
(x,y),(x_test,y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
x.shape
```

```python
def proccess(x,y):
    x = tf.cast(x,dtype=tf.float32)/255.
    y = tf.cast(y,dtype=tf.int32)
    return x,y
```

```python
# 处理喂养数据
```

```python
banch_size = 128
```

```python
train_db = tf.data.Dataset.from_tensor_slices((x,y))
train_db = train_db.map(proccess).shuffle(10000).batch(banch_size)
```

```python
test_db = tf.data.Dataset.from_tensor_slices((x,y))
test_db = test_db.map(proccess).batch(banch_size)
```

```python
# 设置网络
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256,activation='relu'), # [b,784] =>[b,256]
    tf.keras.layers.Dense(128,activation='relu'), # [b,256] =>[b,128]
    tf.keras.layers.Dense(64,activation='relu'), # [b,128] =>[b,64]
    tf.keras.layers.Dense(32,activation='relu'), # [b,64] =>[b,32]
    tf.keras.layers.Dense(10), # [b,32] =>[b,10]
])
```

```python
# 前向传播
```

```python
model.build(input_shape=[None,28*28])
```

```python
model.summary()
```

```python
# 设置优化器
```

```python
optimizer = tf.keras.optimizers.Adam(1e-3)
```

```python
for echop in range(100):
    for step,(x,y) in enumerate(train_db):
        # x dim变换 
        # x:[1,28,28] = >[1,28*28]
        x = tf.reshape(x,[-1,28*28])
        with tf.GradientTape() as tape:
            logits = model(x)
            one_hot_y = tf.one_hot(y,depth=10)

            #损失函数
            loss_mse = tf.reduce_sum(tf.losses.MSE(one_hot_y,logits))
        grads = tape.gradient(loss_mse,model.trainable_variables)
        optimizer.apply_gradients(zip(grads,model.trainable_variables))
        
        if step %100 == 0:
            print(echop,step,"loss:",loss_mse)
    correct_num,total_num = 0,0
    for _,(x,y) in enumerate(test_db):
        # x dim变换 
        # x:[1,28,28] = >[1,28*28]
        x = tf.reshape(x,[-1,28*28])
        logits = model(x)
        prob = tf.nn.softmax(logits,axis=1)
        pred = tf.argmax(prob,axis=1)
        pred = tf.cast(pred,dtype=tf.int32)
        
        correct = tf.cast(tf.equal(y,pred),dtype=tf.int32)
        correct = tf.reduce_sum(correct)
        
        correct_num+=int(correct)
        total_num+=x.shape[0]
    acc = correct_num/total_num
    print("第",echop,"轮:",acc)
```
