```python
# 设置迭代器
```

```python
dbOne = tf.data.Dataset.from_tensor_slices((x,y)).batch(40)
```

```python
type(dbOne)
```

```python
len(dbOne)
```

```python
count = 0
for i in dbOne:
    if count == 20:
        break
    print(i[0])
    count+=1
```

```python
db = iter(dbOne)
```

```python
temp = next(db) # 40个一批
```

```python
# 随机打乱
```

```python
tf.random.shuffle(temp[0])
```

```python
# 或者直接基于原数据集作打乱，也就是基于tensor作打乱操作
```

```python
ok = dbOne.shuffle(100)
```

```python
# 数据预处理
```

```python
# 在拿到数据后先做预处理，设置预处理函数就好了
```

```python
def proccess(x,y):
    x = x/255
    y = y+1
    return x,y
```

```python
# 将banch作为负载体直接进行预处理
```

```python
# 全体Banch作预处理
dbOne.map(proccess)
```
