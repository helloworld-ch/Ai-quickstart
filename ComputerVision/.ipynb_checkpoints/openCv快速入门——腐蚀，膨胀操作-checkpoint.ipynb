```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline

def img_change_chanels(img):
    return cv2.merge(cv2.split(img)[::-1])

def show(img):
    plt.imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def show_sub(images,x,y):
    for i in range(len(images)):
        plt.subplot(x,y,i+1)
        plt.imshow(images[i])
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

### 腐蚀操作  
#### 解释：作用在二分值图像之中，如果255为中心，且在一个卷积核中出现了0，则将该点腐蚀成为0

```python
# 手写腐蚀操作
```

```python
def fushi_img(img,size = 3,iter_num = 1):
    """
    图片，卷积核大小，腐蚀次数
    """
    new_img = np.copy(img)
    padding_size = size//2
    # 填充
    for number in range(iter_num):
        new_img = cv2.copyMakeBorder(new_img,padding_size,padding_size,padding_size,padding_size,cv2.BORDER_CONSTANT,value = 255)
        new_r,new_g,new_b = cv2.split(new_img)
        r,g,b = cv2.split(img)
        for i in range(len(new_r)-size):
            for j in range(len(new_r[0])-size):
                fc = new_r[i:i+size,j:j+size]
                if 0 in fc :
                    r[i][j] = 0
                    g[i][j] = 0
                    b[i][j] = 0
        new_img = cv2.merge([r,g,b])
    
    return new_img
```

```python
def getBinImage(img,thres = 127,max_num = 255):
    indexs = np.mean(img,axis=2)
    chanel = np.copy(img[:,:,0])
    for i in range(len(indexs)):
        for j in range(len(indexs[0])):
            if indexs[i][j]<thres:
                chanel[i][j] = 0
            else:
                chanel[i][j] = max_num
    return cv2.merge([chanel,chanel,chanel])
```

### 膨胀操作
#### 解释： 与腐蚀操作相反，核中有一个255就将它变成255

```python
# 手写膨胀操作
def pengzhang_img(img,size = 3,iter_num = 1):
    """
    图片，卷积核大小，腐蚀次数
    """
    new_img = np.copy(img)
    padding_size = size//2
    # 填充
    for number in range(iter_num):
        new_img = cv2.copyMakeBorder(new_img,padding_size,padding_size,padding_size,padding_size,cv2.BORDER_CONSTANT,value = 255)
        new_r,new_g,new_b = cv2.split(new_img)
        r,g,b = cv2.split(img)
        for i in range(len(new_r)-size):
            for j in range(len(new_r[0])-size):
                fc = new_r[i:i+size,j:j+size]
                if 255 in fc :
                    r[i][j] = 255
                    g[i][j] = 255
                    b[i][j] = 255
        new_img = cv2.merge([r,g,b])
    
    return new_img
```
