```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline

def img_change_chanels(img):
    return cv2.merge(cv2.split(img)[::-1])

def show(img):
    plt.imshow(img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def show_sub(images,x,y):
    for i in range(len(images)):
        plt.subplot(x,y,i+1)
        plt.imshow(images[i])
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
# 灰度图显示
def gray_show(img,name = None):
    if name is not None:
        cv2.imshow(name,img)
    else:
        cv2.imshow("img",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
img = cv2.imread("./my_images/my_img.1.jpg")
img = img_change_chanels(img)
```

```python
show(img)
```

### Sobel算子
### 求梯度的核函数
#### G = Gx+Gy
#### Gx = [[-1,0,1],[-2,0,2],[-1,0,1]]
#### Gy = [[-1,-2,-1],[0,0,0],[1,2,1]]

```python
# api 演示
# Sobel(img,depth,dx,dy,核大小)
# depth 一般为-1即可
# dx 为 使用上述的Gx核
# dy 为 使用上述的Gy核
```

```python
my_img_x = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 3)
my_img_x = cv2.convertScaleAbs(my_img_x)
my_img_y = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 3)
my_img_y = cv2.convertScaleAbs(my_img_y)
my_img = cv2.addWeighted(my_img_x,0.5,my_img_y,0.5,0)
show(my_img)
```

```python
# 灰度图显示
def gray_show(img,name = None):
    if name is not None:
        cv2.imshow(name,img)
    else:
        cv2.imshow("img",img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# 获取边缘
def getBound(img):
    my_img_x = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 3)
    my_img_x = cv2.convertScaleAbs(my_img_x)
    my_img_y = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 3)
    my_img_y = cv2.convertScaleAbs(my_img_y)
    my_img = cv2.addWeighted(my_img_x,0.5,my_img_y,0.5,0)
    return my_img
```

```python
cat_bound = getBound(cat)
show(cat_bound)
```

```python
images = np.hstack([cat,cat_bound])
gray_show(images)
```

```python
temp_bound = getBound(temp)
gray_show(temp_bound)
```
